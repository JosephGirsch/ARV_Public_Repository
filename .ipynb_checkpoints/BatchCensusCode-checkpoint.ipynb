### A. Import Modules

```python
# Census Tract imports
import csv
import censusgeocode as cg 
import pandas as pd
import numpy as np
import datetime as dt
import time
start_time = time.time()

pc = "Silent Gaming" # "SilverJ2"

# Documentation:
# https://pypi.org/project/censusgeocode/
# https://geocoding.geo.census.gov/geocoder/Geocoding_Services_API.pdf
```

```python
def batchGeocode(_df_preBatch, cleanupLoop = False):
    """
    Input: DataFrame. Containing at least these columns: ['Unique ID','Street address','City','State','ZIP']
    Process: Returns the Census Tract and Census Block data for each address, in chunks of 9900. 
    Output: DataFrame. With these columns: id, address, match, matchtype, parsed, tigerlineid, side, statefp, countyfp, tract, block, lat, lon

    _df_preBatch = df_preBatch# DEBUGGING ERRORS
    
    """
    # Correct csvs up until this batch:
    upToHere = 0 #66
    
    subdfSize = 9990 # Max is 10,000. Minimum is 100. 
    subN=0
    _list_geo_dfs = []
    brokenBatches = {}
    file_Path = r"C:\Users\%s\OneDrive\CS 766\project\batchCSVs\\"%pc

    list_dfs = [_df_preBatch[i:i+subdfSize] for i in range(0,_df_preBatch.shape[0],subdfSize)]
    cg_output = cg.CensusGeocode(benchmark = 'Public_AR_Census2020', vintage = 'Census2020_Census2020')
    print("Starting datetime:",dt.datetime.now())
    print("The raw data frame of %d rows is split into %d dataframes of %d length each (%d rows in the last df.)"% (len(_df_preBatch),len(list_dfs), list_dfs[0].shape[0], list_dfs[-1].shape[0]))

    for subDF in list_dfs:
        subN+=1; #print("Starting Batch:",subN,"\tdatetime:",dt.datetime.now())
        ####subDF = subDF.tail(50) 
        ####if subN == 44: # DEBUGGING ERRORS
            ####subDF = subDF.head(150) # DEBUGGING ERRORS
            ###for i in subDF.index:
                ###if i%3==0:
                    ###subDF_trimmed = subDF.loc[i:i+2,['Unique ID','Street address','City','State','ZIP']]; 
                
                
        if (subN <=upToHere)&(cleanupLoop == False):
            print("Batch: %d already saved. Loading... "%subN, end="") 
            file_Path_Load = r"C:\Users\%s\OneDrive\CS 766\project\batchCSVs\postgeo\\"%pc
            file_Path_Load_Name = file_Path_Load +"Debug_PostGeo_Batch" + str(subN)+ ".csv"  
            dfgeo = pd.read_csv(file_Path_Load_Name, index_col = False)
            _list_geo_dfs.append(dfgeo)
            print("\tBatch: %d Loaded "%subN,"datetime:",dt.datetime.now()) 
            
            
        else:
            # If not a cleanup loop, then save to file. 
            file_Path_Name_pre = file_Path + "PreGeo_Batch" + str(subN)+ ".csv" 
            if cleanupLoop == True:
                file_Path_Name_pre = file_Path + "PreGeo_Batch" + str(subN)+ "cleanupLoop" + ".csv"  

            subDF_trimmed = subDF.loc[:,['Unique ID','Street address','City','State','ZIP']]
            subDF_trimmed.to_csv(file_Path_Name_pre, index=False) # Save pre-geo df to csv

            try:
                print("Trying Batch: %d... "%subN, end="") 
                k = cg_output.addressbatch(file_Path_Name_pre) # Obtain post-geo data from censusgeocode

            except:
                print("ERROR. Batch: %d failed. "%subN,"\tdatetime:",dt.datetime.now()) 
                print(subDF_trimmed)
                if subN not in brokenBatches:
                    brokenBatches[subN] = subDF_trimmed

            else:
                print("SUCCESS. Batch: %d correct. "%subN,"\tdatetime:",dt.datetime.now()) 
                dfgeo = pd.DataFrame(k, columns=k[0].keys()).copy() # Convert post-geo data to DataFrame
                
                file_Path_Name_post = file_Path +"\\postgeo\Debug_PostGeo_Batch" + str(subN)+ ".csv"
                if cleanupLoop == True:
                    file_Path_Name_post = file_Path +"\\postgeo\Debug_PostGeo_Batch" + str(subN)+ "cleanupLoop" + ".csv"
                dfgeo.to_csv(file_Path_Name_post, index=False) # Save post-geo df to csv
                _list_geo_dfs.append(dfgeo)
                


    print("\nStarting Merge...","\tdatetime:",dt.datetime.now())
    print("len(_list_geo_dfs):",len(_list_geo_dfs))
    print("Broken batches: ",brokenBatches.keys(),"\n"); 


    _df_geo_merge = pd.concat(_list_geo_dfs, ignore_index=True)     # Given that all the dataframes have the same columns, you can simply concat them.
    _df_geo_merge['match'] = _df_geo_merge['match'].astype('str')     # any bool column with nans has to be converted to 'obj' to match
    _df_geo_merge = _df_geo_merge[_df_geo_merge.id != 'Unique ID']     # remove id row that contains the 'Unique ID' string
    _df_geo_merge['id'] = _df_geo_merge['id'].astype('float').astype('int')     # Change to same type as df_postBatch index now that the 'Unique ID'  string is removed
    _df_geo_merge = _df_geo_merge.set_index('id')     # Replace default index with the carried over 'id' column
    print("Finished Merge.Returning df.","\tdatetime:",dt.datetime.now())
    print("_df_geo_merge.match.value_counts(): ", _df_geo_merge.match.value_counts(),"\n",sep="")
    return _df_geo_merge
```

```python
df = pd.read_csv(r"C:\Users\%s\OneDrive\CS 766\project\MarylandData_preGeo.csv"%pc, keep_default_na=False, dtype={'Levels':'object'})
```

```python
# Step 0: Initialize column changes for prebatch df

batchColumns_inputDict = {'FullStreetAddress':'Street address','City':'City','StateOrProvince':'State','PostalCode':'ZIP'}
batchColumns_inverseDict = {v: k for k, v in batchColumns_inputDict.items()}
outputBatchCols = ['address','match','tract','block']
df_preBatch = df.rename(columns = batchColumns_inputDict)
df_preBatch['Unique ID'] = df_preBatch.index

# Confirm no rows contain the dreaded '"'
if df_preBatch[df_preBatch["Street address"].str.contains('"',na=False)].empty:
    print("Correct! No Address has \" in it!")
else:
    print("Problem! Row has \" in it. Remove these rows! ")
```

```python
%%time
# Step 1: Initialize df_postBatch with first GeoCode Batch
df_geo_merge1 = batchGeocode(df_preBatch) 
df_postBatch = df_preBatch.join(df_geo_merge1[outputBatchCols], how='left', on='Unique ID',lsuffix='_duplicate1', rsuffix='_duplicate2')
print("\n",df_postBatch.match.value_counts(),sep="")
```

```python
print("df_postBatch.match.value_counts(): ",df_postBatch.match.value_counts(),"\n",sep="")

n=4
for i in range(n):
    if 'False' in df_postBatch.match.unique():
        # Step 2: Filter only to rows that failed to match. 
        df_preBatch_loop = df_postBatch.loc[df_postBatch.tract.isna(), :].copy(); print("Does the df_preBatch_loop index and Unique ID still match?:",(df_preBatch_loop['Unique ID'].values == df_preBatch_loop.index.values).all())

        # Step 3: Obtain batchGeocode of only those that failed to match earlier. 
        df_geo_merge_loop = batchGeocode(df_preBatch_loop, cleanupLoop = True) 

        # Step 4: Update postbatch with new data.
        df_postBatch.update(df_geo_merge_loop.loc[:,outputBatchCols], join='left') # update() method directly changes calling object
        print(df_postBatch.match.value_counts(),"\n",sep="")
core_File_Path = r"C:\Users\%s\OneDrive\CS 766\project\batchCSVs\\"%pc
file_Path_Name_postBatch = core_File_Path + "postGeoBatch_Total_" + str(n)+ "_Loops.csv"  
df_postBatch.to_csv(file_Path_Name_postBatch, index=False) # Save post-geo df to csv
```

```python
%%time

benchmarkName = '2020 Census Blocks'  # '2010 Census Blocks'


def getTract(df_gt):
    """
    This code will take in a dataframe and use its 'LATITUDE' and 'LONGITUDE' columns
    to output to the 'tract' column that contains the id of the Census Tract
    DataFrame -> DataFrame
    """
    for row in df_gt.itertuples():
        try:
            cgOutput = cg.coordinates(x=row.Longitude, y=row.Latitude)
            df_gt.loc[row.Index, ['tract']] = cgOutput['Census Tracts'][0]['TRACT']
            print('Success', row.Index, row.FullStreetAddress, row.Latitude, row.Longitude)

        except:
            print('Failed', row.Index, row.FullStreetAddress, row.Latitude, row.Longitude)
            pass

    return df_gt


# Initialize/Confirm existance of required columns
#columnCheck = ['tract']
#for col in columnCheck:
#    if col not in df2.columns:
#        df2[col] = ''

# Continue loop until every row with both Lat and Lon has an accompaning 'tract' value.
count = 0
latlonMask = (df2[['Latitude', 'Longitude']].notna().all('columns')&(df2['Latitude']!=0.0)&(df2['Longitude']!=0.0))
tractMask = (df2['tract']=='')
fullMask = latlonMask & (tractMask)

print("Total remaining rows left to censusGeocode:",len(df2[fullMask]) )
while (len(df2[fullMask]) >= 0) and (count < 12):  # count<10
    df_geoTemp = df2[fullMask].copy()
    df_geoTemp = getTract(df_geoTemp)
    df2[fullMask] = df_geoTemp

    count += 1
    latlonMask = (df2[['Latitude', 'Longitude']].notna().all('columns')&(df2['Latitude']!=0.0)&(df2['Longitude']!=0.0))
    tractMask = (df2['tract']=="")
    fullMask = latlonMask & (tractMask)
    print("lac and lon that are both not NANs:", len(df2[latlonMask]))
    print("Tracts that are NANs:", len(df2[tractMask]))
    print("End Loop:", count, "\n")


#df2.to_csv('MarylandData_Raw_CensusTracts_Febfirst' + '.csv', index=False)
```
